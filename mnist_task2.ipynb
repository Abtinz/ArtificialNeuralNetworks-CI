```python
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import os
```

```python
# function to load images from directory and return them ...
def load_images_from_directory(directory):
    
    images = []
    for filename in os.listdir(directory):
        if filename.endswith('.jpg'):
            image = Image.open(os.path.join(directory, filename))
            image_array = np.asarray(image).reshape(-1)
            images.append(image_array)
            
    return np.array(images)
```

```python
train_images = load_images_from_directory('./datasets/MNIST/2')
print("first image info:")
print(train_images[0])
```

```python
test_images = load_images_from_directory('./datasets/MNIST/5')
```

```python
train_images.shape
```

```python
#we are working with grayscale images which means that we have 8 bit for brightness and maximum number is 255 for normalize
train_images = train_images/255
train_images[0]
```

```python
from layers.convolution2d import Conv2D
from layers.maxpooling2d import MaxPool2D
from layers.fullyconnected import FC
from activations import Activation, get_activation
from optimizers.gradientdescent import GradientDescent
from losses.binarycrossentropy import BinaryCrossEntropy
from model import Model
```

```python
arch = {  
    "C1":Conv2D(in_channels=1, out_channels=32, name="conv1", kernel_size=3, stride=1, padding=1),
    "relu1":get_activation('relu')(),
    "MaxPool1":MaxPool2D(kernel_size=2, stride=2),
    "C2":Conv2D(in_channels=32, out_channels=64, name="conv2", kernel_size=3, stride=1, padding=1),
    "relu2":get_activation('relu')(),
    "MaxPool2":MaxPool2D(kernel_size=2, stride=2),
    "DENSE1":FC(input_size=7*7*64, output_size=128, name="FC1"),
    "s1": get_activation("sigmoid")(),
    "DENSE2":FC(input_size=128, output_size=1, name="FC2"),
    "s2": get_activation("sigmoid")(),
}
```

```python
criterion = BinaryCrossEntropy()
optimizer = GradientDescent(arch, learning_rate=0.01)
```

```python
model = Model(arch, criterion, optimizer)
epocs = 12
savin_url = '/models/mnist52'
```

```python
model.train(train_images,test_images ,epocs, shuffling=True, saving_route = savin_url)
```
